# 2. Familiarising with the file structure

In the project directory azd starter template generated the following assets:
- Infrastructure-as-code (IaC) Terraform modules under the `infra` directory.
- A dev container onfiguration file under the `.devcontainer` directory that installs infrastructure tooling by default. 
- Github and Azure Pipelines CI/CD workflows under the `.github` and the `.azdo` directories that work for most use-cases.
- azure.yaml - A file that describes how the parts of the application map to services in Azure.
- A folder `.azure` containing essential settings for the template's Azure environment, such as the environment name.
